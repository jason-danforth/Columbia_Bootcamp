```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

```python
num_schools = school_data.school_name.count()
```

```python
num_students = student_data.student_name.count()
```

```python
tot_budget = school_data.budget.sum()
```

```python
av_math = round(student_data.math_score.mean(), 2)
```

```python
av_reading = round(student_data.reading_score.mean(), 2)
```

```python
pass_rate = round(((av_math + av_reading) / 2), 2)
```

```python
percent_pass_math = round((student_data.math_score[student_data.math_score >= 70].count() / num_students), 2)
```

```python
percent_pass_reading = round((student_data.reading_score[student_data.reading_score >= 70].count() / num_students), 2)
```

```python
cols = ["num_schools",
        "num_students",
        "tot_budget",
        "av_math",
        "av_reading", 
        "pass_rate", 
        "percent_pass_math", 
        "percent_pass_reading"
       ]

values = [num_schools,
          num_students,
          tot_budget,
          av_math,
          av_reading,
          pass_rate,
          percent_pass_math,
          percent_pass_reading
         ]

df_summary = pd.DataFrame(columns = cols)
df_summary.loc[0] = values
```

```python
df_summary
```

```python
# Format values for clarity

def currency(x):
    return "${:,.0f}".format(x)

df_summary["tot_budget"] = df_summary["tot_budget"].apply(currency)

def commas(x):
    return "{:,.0f}".format(x)

df_summary["num_students"] = df_summary["num_students"].apply(commas)
df_summary["num_schools"] = df_summary["num_schools"].apply(commas)

df_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Create new dataframe to hold to the summary data
# Initial column is name of schools
df_school_summary = pd.DataFrame.from_dict({"name": df.school_name.drop_duplicates().tolist()})
```

```python
# Add new column for school type
df_school_summary['type'] = df[['school_name', 'type']].drop_duplicates()['type'].reset_index(drop=True)
```

```python
# Create a series for the number of students, convert to df, and merge with df_school_summary
num_students_series = df.school_name.value_counts()
df_num_students = pd.DataFrame({'name':num_students_series.index, 'num_students':num_students_series.values})
df_school_summary = df_school_summary.merge(df_num_students)
```

```python
# Add new column for school budget
df_school_summary['budget'] = df[['school_name', 'budget']].drop_duplicates()['budget'].reset_index(drop=True)
```

```python
# Add new column for per-student budget
df_school_summary['per_student_budget'] = df_school_summary['budget'] / df_school_summary['num_students']
```

```python
# Calculate average math score and create new df. Merge new df with school summary df
df_math = df.groupby(['school_name']).math_score.mean().reset_index()
df_math.rename(columns={'school_name':'name', 'math_score':'av_math_score'}, inplace=True)
df_school_summary = df_school_summary.merge(df_math)
```

```python
# Calculate average reading score and create new df. Merge new df with school summary df
df_reading = df.groupby(['school_name']).reading_score.mean().reset_index()
df_reading.rename(columns={'school_name':'name', 'reading_score':'av_reading_score'}, inplace=True)
df_school_summary = df_school_summary.merge(df_reading)
```

```python
# Find # of students passing math, crete dataframe, an merge with df_school_summary
df_passing = df.loc[df.math_score >= 70].groupby(['school_name']).student_name.count().reset_index()
df_passing.rename(columns={'school_name':'name', 'student_name':'passing_math'}, inplace=True)
df_school_summary = df_school_summary.merge(df_passing)

# Calculate % passing math by dividing # passing math by total number of students. Drop # passing math
df_school_summary['% passing math'] = df_school_summary['passing_math'] / df_school_summary['num_students']
df_school_summary.drop(columns=['passing_math'], inplace=True)
```

```python
# Find # of students passing reading, crete dataframe, an merge with df_school_summary
df_passing = df.loc[df.reading_score >= 70].groupby(['school_name']).student_name.count().reset_index()
df_passing.rename(columns={'school_name':'name', 'student_name':'passing_reading'}, inplace=True)
df_school_summary = df_school_summary.merge(df_passing)

# Calculate % passing reading by dividing # passing reading by total number of students. Drop # passing math
df_school_summary['% passing reading'] = df_school_summary['passing_reading'] / df_school_summary['num_students']
df_school_summary.drop(columns=['passing_reading'], inplace=True)
```

```python
# Calculate overall passing rate
df_school_summary['overall passing rate'] = (df_school_summary['% passing math'] + df_school_summary['% passing reading']) / 2
```

## Math Scores by Grade

* Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe

```python
df.head()
```

```python
# Create new df with school name, grade level, and average math score
df_math_grade = df.groupby(['school_name', 'grade']).math_score.mean().reset_index()
```

```python
# Convert grade from int to string
def grade_converter(x):
    if x == "9th":
        x = 9
    elif x == "10th":
        x = 10
    elif x == "11th":
        x = 11
    else:
        x = 12
    return x

df_math_grade.grade = df_math_grade.grade.apply(grade_converter)
```

```python
# Sort values to make df easier to read
df_math_grade = df_math_grade.sort_values(['school_name', 'grade'])
```
