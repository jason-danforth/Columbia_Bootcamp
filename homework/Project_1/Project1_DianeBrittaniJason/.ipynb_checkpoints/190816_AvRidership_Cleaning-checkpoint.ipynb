```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

%matplotlib inline
```

# Build basic dataframe of subway ridership

```python
# Import data
df = pd.read_csv('data\Av_Weekday_Ridership.csv', encoding='latin1')
```

```python
df_ridership = df[df['Station'].notnull()].copy()
```

```python
df_ridership = df_ridership[df_ridership['Station'].str.contains('\*') == False]
```

```python
# Clean data
df_ridership.reset_index(drop=True, inplace=True)
df_ridership.drop(columns=['2018 Rank'], inplace=True)
```

```python
# Convert numeric strings to numbers

def convert_num(item):
    try:
        item = int(item.replace(',', ''))
    except:
        item = item
    return item
```

```python
# Convert percentage strings to numbers

def convert_percent(item):
    try:
        item = float(item.replace('%', ''))
    except:
        item = item
    return item
```

```python
# Make sure all train lines are uppercase

def uppercase(item):
    try:
        item = item.upper()
    except:
        item = item
    return item
```

```python
# Apply changes

df_ridership['Train'] = df_ridership['Train'].apply(uppercase)

df_ridership['2013'] = df_ridership['2013'].apply(convert_num)
df_ridership['2014'] = df_ridership['2014'].apply(convert_num)
df_ridership['2015'] = df_ridership['2015'].apply(convert_num)
df_ridership['2016'] = df_ridership['2016'].apply(convert_num)
df_ridership['2017'] = df_ridership['2017'].apply(convert_num)
df_ridership['2018'] = df_ridership['2018'].apply(convert_num)
```

```python
# Replace NaNs with 0s (all NaNs should be stations that had 0 ridership in 2013)
df_ridership.fillna(0, inplace=True)
```

```python
# Ensure all train names are strings
df_ridership['Train'] = df_ridership['Train'].apply(lambda x: str(x))
```

```python
# Drop stations with bad data 
df_ridership = df_ridership[(df_ridership['2013'] != 0) | (df_ridership['2018'] != 0)]
```

```python
# Create new columns

df_ridership['2018 Change (#)'] = df_ridership['2018'] - df_ridership['2017']
df_ridership['Total Change (#)'] = df_ridership['2018'] - df_ridership['2013']
df_ridership['2016-2018 Change (#)'] = df_ridership['2018'] - df_ridership['2016']

# Percentages go to infinity if denominator is 0. Set max bound at 100%
df_ridership['2018 Change (%)'] = round(((df_ridership['2018'] - df_ridership['2017']) / df_ridership['2017']) * 100, 2)
df_ridership['2018 Change (%)'] = [x if x < 100 else 100 for x in df_ridership['2018 Change (%)']]

df_ridership['Total Change (%)'] = round(((df_ridership['2018'] - df_ridership['2013']) / df_ridership['2013']) * 100, 2)
df_ridership['Total Change (%)'] = [x if x < 100 else 100 for x in df_ridership['Total Change (%)']]

df_ridership['2016-2018 Change (%)'] = round(((df_ridership['2018'] - df_ridership['2016']) / df_ridership['2016']) * 100, 2)
df_ridership['2016-2018 Change (%)'] = [x if x < 100 else 100 for x in df_ridership['2016-2018 Change (%)']]
```

# Overall Ridership

```python
year_1 = df_ridership['2013'].sum()
year_2 = df_ridership['2014'].sum()
year_3 = df_ridership['2015'].sum()
year_4 = df_ridership['2016'].sum()
year_5 = df_ridership['2017'].sum()
year_6 = df_ridership['2018'].sum()

y_values = [year_1, year_2, year_3, year_4, year_5, year_6]
x_values = [0, 1, 2, 3, 4, 5]
labels = ['2013', '2014', '2015', '2016', '2017', '2018']

fig, ax = plt.subplots(figsize=(10,5))
plt.plot(x_values, y_values, marker='o', 
         markersize=9, markerfacecolor='w', markeredgecolor='#0039A6', markeredgewidth=3, color='#0039A6')

ax.set_facecolor('#F2F1F0')
plt.xlabel("Years")
plt.xticks(x_values, labels)
plt.ylabel('Total Ridership')
plt.title('Total Subway Ridership by Year')
plt.grid(True)

plt.savefig('images\Total_Ridership.png')
```

# Compare Train Lines

```python
names = ['A', 'C', 'E', 'B', 'D', 'F', 'M', 'G', 'J', 'Z', 'L', 'N', 'Q', 'R', 'W', 'S', '1', '2', '3', '4', '5', '6', '7']
```

```python
#Get colors
df_colors = pd.read_csv('data\subway_colors.csv')

colors = []
for name in names:
    colors.append(df_colors[df_colors['Line'] == name]['Color'].tolist()[0])
```

```python
# Create list of values (total inc/dec in ridership and average % inc/dec in ridership) for each train
total_inc = []
mean_inc = []

for name in names:
    total_increase = df_ridership[df_ridership['Train'].str.contains(name)]['Total Change (#)'].sum()
    total_ridership = df_ridership[df_ridership['Train'].str.contains(name)]['2018'].sum()
    
    total_inc.append(total_increase)
    mean_inc.append(round(total_increase/total_ridership, 2))
```

```python
y_values = total_inc
x_values = [x for x in range(len(names))]

fig, ax = plt.subplots(figsize=(10,5))

for i in range(len(y_values)):
    plt.plot(x_values[i], y_values[i], linestyle='None', 
             marker='o', markersize=9, markerfacecolor='w', markeredgecolor=colors[i], markeredgewidth=3)

plt.grid(True)
ax.set_facecolor('#F2F1F0')
plt.hlines(0, min(x_values)-1, max(x_values)+1, colors='k', linestyles='solid', linewidth=1)
plt.xticks(x_values, names)
plt.xlabel('Subway Line')
plt.ylabel('Av. Daily Increase/Decrease')
plt.title('Change in Ridership by Subway Line 2013-2018');

plt.savefig('images\Change_per_line.png')
```

# Scatter plot for all stops

```python
#Set variables
x_axis = '2018'
y_axis = '2016-2018 Change (%)'
size = '2016-2018 Change (#)'
```

```python
#Remap sizes
size_og = abs(df_ridership[size])
min_og = min(size_og)
max_og = max(size_og)

min_new = 20
max_new = 1000
size_new = (((size_og - min_og) * (max_new - min_new)) / (max_og - min_og)) + min_new
```

```python
#Set colors by borough
boroughs = df_ridership['Borough']

colors = []
for borough in boroughs:
    if borough == 'The Bronx':
        colors.append('y')
    elif borough == 'Brooklyn':
        colors.append('b')
    elif borough == 'Queens':
        colors.append('m')
    else:
        colors.append('c')
```

```python
#Plot
x_val = df_ridership[x_axis]
y_val = df_ridership[y_axis]

fig, ax = plt.subplots(figsize=(20,10))

plt.scatter(x_val, y_val, alpha=0.5, color=colors, s=size_new)

plt.grid(True)
ax.set_facecolor('#F2F1F0')
plt.xlabel(f'Total ridership, {x_axis}')
plt.ylabel(y_axis)
plt.title('System-Wdie Change in Ridership, 2016-2018')
```
